```python
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')

import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('home_data.csv')
sales
```

```python
%matplotlib inline

sales_plot = sales.plot(x = 'sqft_living', y='price', kind='scatter')
plt.title('House prices')
plt.xlabel('Area in square feet')
plt.ylabel('Price')
```
